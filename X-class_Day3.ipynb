Playing with a Random Forest Classifier of an entire event list. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import copy
import seaborn as sns
import matplotlib.cm as cm
from astropy.table import Table
from sklearn.ensemble import RandomForestClassifier
sns.set(style="whitegrid")
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

%matplotlib inline
```

Read in the event lists

```python
e = Table.read('Data/evt_1229.fits',hdu=1) # full event list
s = Table.read('Data/sevt_1229.fits',hdu=1) # event list for sources 
b = Table.read('Data/bevt_1229.fits',hdu=1) # event list for background 

# copies to remove columns
ec = copy.copy(e) 
sc = copy.copy(s) 
bc = copy.copy(b) 
```

Some columns are not helpful at this stage, some are not compatible with pandas (i.e., status). Consider editting this list of bad columns. 

```python
# remove problematic and redundant columns
badcols = ['status','x','y','ccd_id','expno','node_id','chipx','chipy','tdetx','tdety','detx','dety'] # ,'pi','pha','pha_ro']

for bcol in badcols: 
    del sc[bcol]
    del bc[bcol]
    del ec[bcol]

# convert to pandas data frames
sd = sc.to_pandas()
bd = bc.to_pandas()
ed = ec.to_pandas()
```

```python
training_data = pd.concat([sd[np.arange(len(sd))%2==0],bd[np.arange(len(bd))%2==0]])
training_labels  = np.zeros(len(training_data))
training_labels[0:len(range(0,len(sd),2))] = 1.0
```

```python
testing_data = pd.concat([sd[np.arange(len(sd))%2==1],bd[np.arange(len(bd))%2==1]])
testing_labels = np.zeros(len(testing_data))
testing_labels[0:len(range(1,len(sd),2))] = 1.0
```

```python
rfc = RandomForestClassifier(n_estimators=200,oob_score=True)
rfc.fit(training_data.values,training_labels)
```

```python
print "OOB Score: {0}".format(rfc.oob_score_)
print sorted(zip(training_data.columns.values,rfc.feature_importances_),key=lambda q: q[1],reverse=True)
```

```python
testing_result = rfc.predict(testing_data)
```
